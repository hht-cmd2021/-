```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.font_manager import _rebuild
_rebuild()
mpl.rcParams["font.sans-serif"] = [u"SimHei"]
mpl.rcParams["axes.unicode_minus"] = False
```

```python
#纵向柱状图   bar()
def fun1():
    #不同城市房价信息对比
    price = [90000,92000,70000,85000]
    #第一个参数表示x轴的位置，其实也是表示有几根柱子，第二个参数是y轴的值，也是具体的数据的值
    #width:柱的宽度
    #color:柱的颜色
    #alpha:透明度
    plt.bar(np.arange(4), price,width=0.5,color="steelblue",alpha=0.8)
    #修改刻度，将当前x轴的刻度修改为对应的城市名
    plt.xticks(np.arange(4),["北京","上海","广州","深圳"])
    #添加标题
    plt.title("一线城市房价")
fun1()
```

```python
#横向的柱状图  barh
def fun2():
    #不同城市房价信息对比
    price = [90000,92000,70000,85000]
    #第一个参数表示y的位置，第二个参数是数据，表示x轴的位置
    plt.barh(np.arange(4),price, height=0.5, color="steelblue")
    #修改刻度
    plt.yticks(np.arange(4),["北京","上海","广州","深圳"])
    
    #将x轴的范围变大
    plt.xlim([0,110000])
    
    #在每一根柱的后面添加具体的房价
    for x,y in zip(np.arange(4),price):
        plt.text(y,x,y)
fun2()
```

```python
#两组柱子对比图
def fun3():
    #2017年各城市房价
    y2017 = [90000,92000,70000,85000]
    #2018年各城市房价
    y2018 = [110000,100000,85000,93000]
    #各城市名称
    label = ["北京","上海","广州","深圳"]
    
    #柱的宽度
    bar_width = 0.3
    
    #画第一组柱子
    plt.bar(np.arange(4), y2017,color="steelblue",alpha=0.8,width=bar_width, label="2017")
    #画第二组柱子,第二组柱子要在第一组柱子的右侧紧挨着，所以要加bar_width
    plt.bar(np.arange(4)+bar_width,y2018, color="indianred", alpha=0.8,width=bar_width, label=2018)
    #显示图例
    plt.legend()
    #修改刻度
    plt.xticks([0.15,1.15,2.15,3.15],label)
    plt.ylim([0,130000])
    
    for x,y in zip(np.arange(4), y2017):
        plt.text(x-0.15,y+1000,y)
    
    for x,y in zip(np.arange(4), y2018):
        plt.text(x+bar_width-0.15,y+1000,y)
fun3()
```

```python
#饼图 pie
def fun4():
    #不同学历的就业率
    edu = [0.2515,0.3724,0.3336,0.0368,0.0057]
    #不同的学历名称
    label = ["中专","大专","本科","硕士","其它"]
    #不同的颜色
    color = ["#9999ff","#ff9999","#7777aa","#2442aa","#99ffcc"]
    
    #用来突出显示大专
    ex = [0,0.1,0,0,0]
    
    #如果遇到不是正圆的情况，可以调用下列代码
    plt.axes(aspect="equal")
    
    #修改x、y轴的取值范围
    plt.xlim([0,4])
    plt.ylim([0,4])
    
    plt.pie(x=edu,#数据
           labels=label,#每一块饼的标注
           colors=color,#每块饼的颜色
           autopct="%.1f%%",#显示占比
           frame=True,#显示画板边框，默认是不显示的
           center=(2,2),#修改饼的圆心点
           radius=1.5,#半径
           startangle=180,#旋转180度
           explode=ex)
fun4()
```

```python
#直方图  通常用来查看一组数据中每一种内容出现的频率 hist()
def fun5():
    #x数据
    #orientation方向 默认是vertical,  还可以设置horizontal
    #plt.hist(x)
    plt.hist(x, orientation="horizontal")
fun5()
```

```python
#散点图  scatter()
def fun6():
    #用来确定点的位置坐标（x,y）
    x = np.random.randn(1000)
    y = np.random.randn(1000)
    
    #该函数需要绘制1000个随机颜色的点
    #生成3000个随机数，调用reshape方法，将该数组的的结构改成（1000，3）,相当于有1000组数，每组数有3个数值，这3个数值就可以表示一种颜色
    colors = np.random.random(3000).reshape((1000,3))
    
    #画散点图
    #x,y表示点的坐标
    #color颜色
    #s 点的大小
    plt.scatter(x,y,color=colors, s=10)
      
fun6() 
```

```python
#一个画板中可以画多个图，这里就要涉及到子图的概念
def fun7():
    plt.figure(figsize=(12,8))
    #如果想画子图，需要调用subplot(a,b,c)  a行  b列   c第几张图
    #第1行第1列
    a1 = plt.subplot(2,3,1)
    x1 = np.arange(5)
    y1 = x1 + 5
    a1.plot(x1,y1)
    a1.grid(color="r", linestyle="--")
    
    #第1行第2列
    a2 = plt.subplot(2,3,2)
    x2 = np.linspace(-np.pi,np.pi,256)
    y2 = np.sin(x2)
    a2.plot(x2,y2)
    a2.grid(color="g", linestyle="-.")
    
    #第1行第3列
    a3 = plt.subplot(2,3,3)
    x3 = np.linspace(-np.pi, np.pi,256)
    y3 = np.cos(x3)
    a3.plot(x3,y3)
    a3.grid(color="b", linestyle=":")
    
    #第2行第1列
    a4 = plt.subplot(2,3,4)
    x4 = np.arange(5)
    y4 = x4 + 5
    a4.plot(x4,y4)
    a4.grid(color="r", linestyle="--")
    
    #第2行第2列
    a5 = plt.subplot(2,3,5)
    x5 = np.linspace(-np.pi,np.pi,256)
    y5 = np.sin(x5)
    a5.plot(x5,y5)
    a5.grid(color="g", linestyle="-.")
    
    #第2行第3列
    a6 = plt.subplot(2,3,6)
    x6 = np.linspace(-np.pi, np.pi,256)
    y6 = np.cos(x6)
    a6.plot(x6,y6)
    a6.grid(color="b", linestyle=":")

fun7()
```

```python
#图片保存
#图片保存前一定要先调用代码创建图片,然后再执行保存图片的代码
fun7()
#fname是保存图片的路径和名字，如果没有写路径，默认保存在当前文件的目录下
#dpi 分辨率，每英寸上的点数，分辨率越大，图片放大时越清晰
plt.savefig(fname="pic.jpg",dpi=500)
```
