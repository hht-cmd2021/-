```python
#处理data2中带有json字符串的列，将其转换成python可以访问的对象
json_col = ["cast","crew"]
for column_name in json_col:
    data2[column_name] = data2[column_name].apply(json.loads)
```

```python
#join函数复习
l = ["a","b","c"]
s = ",".join(l)
s
```

```python
#处理cast列，该列中存放的是每个演员的信息，只保留姓名
#一部电影中多个演员的姓名用逗号分隔
def getActorName(x):
    l = []
    #x表示某部电影中所有的演员信息
    for actor in x:
        #actor表示某部电影中某一个演员的信息
        #将演员的姓名取出放入列表中
        l.append(actor["name"])
        #将当前电影中所有演员的姓名取出，用逗号连接
    return ",".join(l)
        
#data2["cast"] = data2["cast"].apply(getActorName)
data2["cast"] = data2["cast"].\
    apply(lambda x : ",".join([actor["name"] for actor in x]))
```

```python
#处理crew列，只保留导演的姓名，导演是director
data2["crew"] = data2["crew"].\
    apply(lambda x : ",".join([i["name"] for i in x if(i["job"]=="Director")]))
```

```python
data2.head(2)
```

```python
#将列名crew改为director
data2.rename(columns={"crew":"director"},inplace=True)
```

```python
data2.head(2)
```

```python
#将data1中所有带json字符串的内容转换成python对象
#genres, keywords, production_companies
#production_countries, spoken_languages
json_col = ["genres","keywords","production_companies","production_countries","spoken_languages"]
for i in json_col:
    data1[i] = data1[i].apply(json.loads)
```

```python
def getNames(x):
    return ",".join([i["name"] for i in x])
data1["genres"] = data1["genres"].apply(getNames)
data1["keywords"] = data1["keywords"].apply(getNames)
data1["production_companies"] = data1["production_companies"].apply(getNames)
data1["production_countries"] = data1["production_countries"].apply(getNames)
data1["spoken_languages"] = data1["spoken_languages"].apply(getNames)
```

```python
data1.head(2)
```

```python
#删除data1中的无用列title，因为data2中也有
del data1["title"]
```

```python
#合并两个数据集
#由于数据行数一致，猜测是一对一的，可以使用内连接
df = pd.merge(data2,data1,how="inner",left_on="movie_id",right_on="id")
```

```python
df.head(2)
```

```python
#删除无用列
del df["id"]
del df["homepage"]
del df["original_title"]
del df["overview"]
del df["spoken_languages"]
del df["tagline"]
```

```python
#获取当前数据集中所有的电影类型
types = set()
for item in df.genres.str.split(","):
    #item表示某一部电影的所有类型，数据类型是list
    for i in item:
        types.add(i)
movie_types = list(types)
```

```python
#发现第一个位置是空的，删除
movie_types.pop(0)
```

```python
#普通形式
def fun1(x):
    if(x):
        return 1
    else:
        return 0
fun1(True)
```

```python
#if表达式形式
def fun2(x):
    return 1 if x else 0
fun2(True)
```

```python
for i in movie_types:
    df[i] = df.genres.str.contains(i).\
        apply(lambda x : 1 if x else 0)
```

```python
df.head(2)
```

```python
#从release_date中提取年份和月份，并新增两列
#release_year   release_month
df["release_year"] = pd.to_datetime(df.release_date,format="%Y-%m-%d").dt.year
df["release_month"] = pd.to_datetime(df.release_date,format="%Y-%m-%d").dt.month
```

```python
df.head(2)
```

```python
#不同年份，有多少动作片上映
#根据年份进行分组，然后在组内对Action列进行统计
result1 = df.groupby("release_year")["Action"].count()
result1
```

```python
plt.figure(figsize=(12,8))
plt.plot(result1.index,result1,"-")
plt.xticks(np.arange(1915,2018,5))
plt.xlim([1915,2016])
```

```python
#每年每种类型电影的产量
#为了方便画图，这里将release_year列做为索引
df.index = df.release_year
```

```python
#这里只需要年份和类型，不需要其它列
df_type = df[movie_types]
df_type.head(2)
```

```python
#进一步处理数据，进行分组求和，获取不同年份每种类型电影的产量
df_type2 = df_type.groupby("release_year").sum().sort_index(ascending=False)
df_type2
```

```python
plt.figure(figsize=(12,8))
plt.plot(df_type2,label=df_type2.columns)
plt.legend(df_type2)
plt.xticks(np.arange(1915,2018,5))
plt.xlim([1915,2016])
```
