```python
#声明一个函数，可以实现动态创建DataFrame
def mk_df(cols,inds):
    """
        cols：列
        inds:索引
    """
    data = {c:[c+str(i) for i in inds] for c in cols}
    return DataFrame(data, columns=cols, index=inds)
```

```python
mk_df(["A","B"],["1","2"])
```

```python
#pandas支持使用concat来进行数据合并
d1 = mk_df(["A","B"],["0","1"])
d2 = mk_df(["A","B"],["2","3"])
display(d1,d2)
```

```python
#pd.concat(['objs', 'axis=0', "join='outer'", 'join_axes=None', 'ignore_index=False', 'keys=None', 'levels=None', 'names=None', 'verify_integrity=False', 'sort=None', 'copy
pd.concat((d1,d2))
    
```

```python
#可以通过设置axis=1改变合并方向，但这么干没意义
pd.concat((d1,d2), axis=1)
```

```python
d3 = mk_df(["A","B"],["0","1","2"])
d4 = mk_df(["A","B"],["1","2","3"])
display(d3,d4)
```

```python
#合并时，索引可以重复
pd.concat((d3,d4))
```

```python
#如果对索引不满意，可以在合并时重新分配索引，要设置ignore_index=True
pd.concat((d3,d4),ignore_index=True)
```

```python
#以上的列名都是一致的，属性匹配合并
#下面看一下不匹配的
#1.外连接，补NaN（默认）  数据库中叫full outer join
d1 = mk_df(["A","B"],[1,3])
d2 = mk_df(["B","C"],[2,4])
display(d1,d2)
```

```python
pd.concat((d1,d2), join="outer") #默认join是outer
```

```python
#2.内连接  数据库中叫：内连接，等值连接
pd.concat((d1,d2), join="inner")
```

```python
d3 = mk_df(["A","C","D"],[0,2,4])
d4 = mk_df(["C","D","E"],[3,6,9])
display(d3,d4)
```

```python
pd.concat((d3,d4))
```

```python
#3.join_axes属性可以指定以哪个DataFrame的列为准进行数据合并，类似数据库中的左、右连接
pd.concat((d3,d4),join_axes=[d3.columns])
```

```python
pd.concat((d3,d4),join_axes=[d4.columns])
```

```python
d1 = DataFrame({"employ":["Sheldon","Penny","Lanard"],"dept":["hr","marketing","sales"],"salary":[10000,9000,8500]})
d2 = DataFrame({"employee":["Sheldon","Penny","Kite"],"dept":["hr","marketing","sales"],"work_time":[2,4,5]})
display(d1,d2)
```

```python
#左右指的是参数顺序，下列代码中，d1为左表，d2为右表
#下面的代码是：以d1中的employ和d2中的employee两列进行匹配，合并数据
pd.merge(d1,d2,left_on="employ", right_on="employee")
```

```python
#how属性表示的是连接方式，默认是inner，内连接
#可以修改成outer 外连接
pd.merge(d1,d2,left_on="employ", right_on="employee", how="outer")
```

```python
#还可以修改成left，左连接
pd.merge(d1,d2,left_on="employ", right_on="employee", how="left")
```

```python
#还可以修改成right,右连接
pd.merge(d1,d2,left_on="employ", right_on="employee", how="right")
```

```python
#同样使用左连接，只要把DataFrame参数的顺序调换一下，就可以实现上面的右连接效果
pd.merge(d2,d1,left_on="employee", right_on="employ", how="left")
```
