```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt

#支持中文
import matplotlib as mpl
from matplotlib.font_manager import _rebuild
_rebuild()
mpl.rcParams["font.sans-serif"] = [u"SimHei"]
mpl.rcParams["axes.unicode_minus"] = False
```

```python
#线图
def fun1():
    x = np.arange(10) #x轴所有的点
    y = x #y轴所有的点
    
    #获取画板  如果不需要设置画板大小，可以省略下面一行代码
    plt.figure()
    
    #画线
    #plt.plot(x,y) #默认线
    #plt.plot(x,y,"-x") #加入了线的样式
    plt.plot(x,y,"-x", color="m")

fun1()
#?plt.plot
```

```python
def fun2():
    x = np.arange(10)
    y = x ** 2
    plt.figure()
    plt.plot(x, y, "--", color="b")

fun2()
```

```python
def fun3():
    #从1到10之间，取100个数，这100个数是等差的
    x = np.linspace(1,10,100)
    y = np.cos(x)
    z = np.sin(x)
    plt.figure()
    plt.plot(x, y, "-2")
    plt.plot(x, z, "-2")
fun3()
```

```python
def fun4():
    x = np.arange(10)
    y = x ** 2
    #设置画板宽8英寸，高4英寸
    plt.figure(figsize=(8,4))
    plt.plot(x, y, "-x", color="r")
    
    #为x,y轴分别添加标注，默认不支持中文
    #plt.xlabel("time")
    #plt.ylabel("salary")
    plt.xlabel("时间")
    plt.ylabel("收入")
    
    #修改刻度
    plt.xticks(np.arange(10))
    #plt.xticks(np.arange(10),["a","b","c","d","e","f","g","h","i","j"])
    
    for i, j in zip(x,y):
        #text()方法可以用来在画板上添加文字
        #plt.text(x, y, s, fontdict=None, withdash=False, **kwargs)
        #x:要添加文字的x轴坐标
        #y:要添加文字的y轴坐标
        #s:要添加的文字内容
        #bbox给文字添加一个背景块，颜色是red，透明度0.5
        plt.text(i+0.2,j,j,bbox=dict(facecolor='red', alpha=0.5))
    
    #给当前界面添加表格
    plt.grid(True)
fun4()
```

```python
def fun5():
    
    x = np.arange(50)
    
    #列表中存放6条线的不同颜色
    colors = ["r","y","g","b","c","m"]
    #字典中存放6条线的不同样式
    kind = {"circle":"o","start":"*","traggle_down":"v","dash_dot":"-","dash_line":"--","x-mark":"x"}
    
    plt.figure(figsize=(12,8))
    
    for index,key in zip(np.arange(6),kind.keys()):
        #产生50个随机数，并且这些随机数是符合正态分布的
        y = np.random.randn(50)
        plt.plot(x,y,kind[key], color=colors[index], label=key)
    
    #显示图例
    #loc参数表示位置，默认位置是best 0 ，也就是最佳位置
    plt.legend(loc=0)
fun5()
```

```python
def fun6():
    plt.figure(figsize=(8,6))
    #np.pi就是派
    x = np.linspace(-np.pi,np.pi,256,endpoint=True)
    c = np.cos(x)
    s = np.sin(x)
    
    plt.plot(x, c, "-")
    plt.plot(x, s, "--")
    
    #指定x轴、y轴的取值范围
    plt.xlim([-4,4])
    plt.ylim([-1,1])
    
    #修改x轴、y轴刻度
    plt.yticks([-1,0,1])
    #r"$内容$" 表示将内容变成斜体  \pi表示pai的转义字符
    plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],[r"$-\pi$",r"$-\pi/2$",r"$0$",r"$\pi/2$",r"$\pi$"])
    
    #直角坐标系只需要两条线，去除上边框和右边框
    #通过plt.gca()获取当前坐标轴
    ax = plt.gca() #gca = get current axis
    ax.spines["right"].set_color("none")
    ax.spines["top"].set_color("none")
    
    #将底边框移到y轴0点位置
    ax.spines["bottom"].set_position(("data",0))
    
    #将左边框移到x轴0点位置
    ax.spines["left"].set_position(("data",0))
fun6()
```
